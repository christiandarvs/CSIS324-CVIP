```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("./images/medical-image.jpg")
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Loaded Image")
plt.axis("off")
plt.show()
```

### Part 2: Basic Image Processing

```python
# This code converts an image to grayscale, resizes it to 300x300 pixels, saves both the grayscale and resized images, and then displays them side by side using Matplotlib.

gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
resized_img = cv.resize(gray_img, (300, 300))

cv.imwrite("gray_image.jpg", gray_img)
cv.imwrite("resized_image.jpg", resized_img)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(gray_img, cmap="gray")
axes[0].set_title("Grayscale Image")
axes[0].axis("off")

axes[1].imshow(resized_img, cmap="gray")
axes[1].set_title("Resized Image (300 x 300)")
axes[1].axis("off")
plt.show()
```

```python
if img is None:
    print("Error: Image not found!")
else:
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    blurred = cv.GaussianBlur(gray, (5, 5), 0)

    plt.Figure(figsize=(6, 6))
    plt.imshow(blurred, cmap="gray")
    plt.title("Blurred Grayscale Image")
    plt.axis("off")
    plt.show()
```

### Part 3: Edge Detection using the Canny Method

```python
# Global Thresholding
ret, thresh_global = cv.threshold(blurred, 127, 255, cv.THRESH_BINARY)

# Adaptive Thresholding (Mean Adaptive)
thresh_adaptive = cv.adaptiveThreshold(
    blurred, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2
)

# Display segmentation results
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(thresh_global, cmap="gray")
axes[0].set_title("Global Thresholding")
axes[0].axis("off")

axes[1].imshow(thresh_adaptive, cmap="gray")
axes[1].set_title("Adaptive Thresholding")
axes[1].axis("off")
plt.show()
```

### Part 4: Edge Detection using the Canny Method

```python
edges = cv.Canny(blurred, 50, 150)

plt.figure(figsize=(6, 6))
plt.imshow(edges, cmap="gray")
plt.title("Canny Edge Detection")
plt.axis("off")
plt.show()
```

```python
edges = cv.Laplacian(blurred, 50, 150)

plt.figure(figsize=(6, 6))
plt.imshow(edges, cmap="gray")
plt.title("Laplacian Edge Detection")
plt.axis("off")
plt.show()
```

```python
sobel_x = cv.Sobel(blurred, cv.CV_64F, 1, 0, ksize=3)  # Sobel X
sobel_y = cv.Sobel(blurred, cv.CV_64F, 0, 1, ksize=3)  # Sobel Y
sobel_combined = cv.magnitude(sobel_x, sobel_y)  # Combine both directions

# Display the result
plt.figure(figsize=(6, 6))
plt.imshow(sobel_combined, cmap="gray")
plt.title("Sobel Edge Detection")
plt.axis("off")
plt.show()
```

### Part 5: Combining Segmentation and Edge Detection

```python
thresh_color = cv.cvtColor(thresh_global, cv.COLOR_GRAY2BGR)

overlay = thresh_color.copy()
overlay[edges != 0] = [255, 0, 0]

plt.figure(figsize=(6, 6))
plt.imshow(cv.cvtColor(overlay, cv.COLOR_BGR2RGB))
plt.title("Overlay of Edges on Segmented Image")
plt.axis()
plt.show()
```

```python
if img is None:
    print("Error: Image not found!")
else:
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

    # Define different Gaussian blur kernel sizes
    kernel_sizes = [(3, 3), (5, 5), (7, 7)]

    # Define different threshold values
    threshold_values = [100, 127, 150]

    fig, axes = plt.subplots(len(kernel_sizes), len(threshold_values), figsize=(12, 9))

    for i, ksize in enumerate(kernel_sizes):
        blurred = cv.GaussianBlur(gray, ksize, 0)

        for j, thresh in enumerate(threshold_values):
            _, binary_thresh = cv.threshold(blurred, thresh, 255, cv.THRESH_BINARY)
            axes[i, j].imshow(binary_thresh, cmap="gray")
            axes[i, j].set_title(f"Kernel: {ksize}, Thresh: {thresh}")
            axes[i, j].axis("off")

    plt.tight_layout()
    plt.show()
```
